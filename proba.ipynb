## Rating

```python
df_r = pd.read_parquet("data/complete/rating.parquet")
df_r.head()
```

```python
df_r = df_r.reset_index()
```

```python
df_r.head()
```

```python
df_r = df_r.reset_index().rename(columns={"index": "rating_id"}).set_index("rating_id")
```

```python
df_r = df_r[~df_r["rating"].isnull()]
```

```python
df_r.head()
```

```python
df_r = df_r.reset_index().set_index(["bg_id__bg_id", "user_name__user_id", "date"])
```

```python
df_r = df_r.drop(["index"], axis = 1)
```

```python
df_r = df_r.drop_duplicates()
```

```python
df_r.reset_index()
```

```python
df_r.index.value_counts()
```

```python
df_r.query("bg_id__bg_id == 150145 and user_name__user_id == 'Zzaga'")
```

```python
df_r = df_r.reset_index().rename(columns= { "user_id__user_name": "user_name__user_id"})
```

```python
df_r.to_parquet("data/complete/rating.parquet")
```
